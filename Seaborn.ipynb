```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
sns.pairplot(tips,hue="sex",palette="Blues")
```

```python
Distribution Plot
```

```python
crash = sns.load_dataset('car_crashes')
sns.displot(crash['not_distracted'],kde=False,bins=25)
```

```python
Joint plot
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash,kind='kde')
```

```python
KDE plot kernal density estimation
```

```python
sns.kdeplot(crash['alcohol'])
```

```python
Pair Plot
```

```python
sns.pairplot(crash)
```

```python
Rug plot
```

```python
sns.rugplot(tips["tip"])
```

```python
crash = sns.load_dataset('car_crashes')
sns.set_style('ticks')
plt.figure(figsize=(4,4))
sns.set_context('paper',font_scale=1.4)
sns.jointplot(x='speeding',y='alcohol',data= crash,kind='reg')
```

```python
Categorical plots
```

```python
Bar plots
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
count Plot
```

```python
sns.countplot(x='sex',data=tips)
```

```python
BOX plot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
plt.legend(loc=0)
```

```python
Violin Plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
crash = sns.load_dataset('car_crashes')
crash_new=crash.drop(columns=['abbrev'],inplace=True)
```

```python
flights=sns.load_dataset('flights')
flights=flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(flights,cmap='Blues',linecolor='white',linewidth=1)
```
